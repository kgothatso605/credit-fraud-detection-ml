# Univariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
from pathlib import Path

# Find project root and set working directory
current = Path.cwd()
while current.name != "credit_fraud_ml":
    current = current.parent
os.chdir(current)
sys.path.append(str(current))

from src.data.data_loader import DataLoader

loader = DataLoader()
data = loader.load_all('train')
transactions = data['transactions']
customers = data['customers']
terminals = data['terminals']
```

```python
numeric_cols = transactions.select_dtypes(include=np.number).columns
numeric_cols
```

```python
for col in numeric_cols:
    plt.figure(figsize=(10,4))
    sns.histplot(transactions[col], bins=50)
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(10,3))
    sns.boxplot(x=transactions[col])
    plt.title(f'{col} Boxplot')
    plt.show()
```
